BANGALORE CARS

```python
bangalore
```

```python
bangalore['new_car_detail'] = bangalore['new_car_detail'].apply(ast.literal_eval)
bangalore['new_car_overview'] = bangalore['new_car_overview'].apply(ast.literal_eval)
bangalore['new_car_feature'] = bangalore['new_car_feature'].apply(ast.literal_eval)
bangalore['new_car_specs'] = bangalore['new_car_specs'].apply(ast.literal_eval)
```

```python
bangalore['Fuel_Type'] = bangalore['new_car_detail'].apply(lambda x: x.get('ft', ''))
bangalore['Body_Type'] = bangalore['new_car_detail'].apply(lambda x: x.get('bt', ''))
bangalore['Mileage_km'] = bangalore['new_car_detail'].apply(lambda x: x.get('km', ''))
bangalore['Transmission type']= bangalore['new_car_detail'].apply(lambda x: x.get('transmission',''))
bangalore['Number owner']= bangalore['new_car_detail'].apply(lambda x: x.get('ownerNo',''))
bangalore['oem'] =  bangalore['new_car_detail'].apply(lambda x: x.get('oem',''))
bangalore['Car model'] =  bangalore['new_car_detail'].apply(lambda x: x.get('model',''))
bangalore['modelYear'] =bangalore['new_car_detail'].apply(lambda x: x.get('modelYear',''))
bangalore['centralVariantId' ]= bangalore['new_car_detail'].apply(lambda x: x.get('centralVariantId',''))
bangalore['variantName'] = bangalore['new_car_detail'].apply(lambda x: x.get('variantName',''))
bangalore['price'] = bangalore['new_car_detail'].apply(lambda x: x.get('price',''))
```

```python
bangalore['Overview_heading'] = bangalore['new_car_overview'].apply(lambda x: x.get('heading', ''))
bangalore['Top_key'] = bangalore['new_car_overview'].apply(lambda x: x.get('top', []))
```

```python
bangalore['Top_key_value'] = bangalore['Top_key'].apply(lambda x: x[0]['value'] if x and isinstance(x, list) and len(x) > 0 else '')
```

```python
bangalore['Features'] = bangalore['new_car_specs'].apply(lambda x: x.get('heading', ''))
bangalore['Top specification'] = bangalore['new_car_specs'].apply(lambda x: x.get('top', []))
bangalore['Detailed engine'] = bangalore['new_car_specs'].apply(lambda x: x.get('data', []))
```

```python
bangalore.drop(['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs', 'car_links','Overview_heading','Top_key','Features'],axis = 1,inplace = True)
```

```python
selected_features = bangalore[['Top specification', 'Detailed engine']].copy()
selected_features
```

```python
def extract_features(row):
    features = {
        'Mileage': None,
        'Engine': None,
        'Max Power': None,
        'Torque': None,
        'Seats': None
    }

    if isinstance(row, list):
        for item in row:
            if 'key' in item and 'value' in item:
                key, value = item['key'], item['value']
                if key in features:
                    features[key] = value

    return features

features_df_b = selected_features['Top specification'].apply(extract_features).apply(pd.Series)
```

```python
features_df_b
```

```python
features_df_b.isnull().sum()
```

```python
mileage_mode = features_df_b['Mileage'].mode()[0]
Engine_mode = features_df_b['Engine'].mode()[0]
Max_Power_mode = features_df_b['Max Power'].mode()[0]
Torque_mode = features_df_b['Torque'].mode()[0]
Seats_mode = features_df_b['Seats'].mode()[0]
```

```python
features_df_b['Mileage'] = features_df_b['Mileage'].fillna(mileage_mode)
features_df_b['Engine'] = features_df_b['Engine'].fillna(Engine_mode)
features_df_b['Max Power'] = features_df_b['Max Power'].fillna(Max_Power_mode)
features_df_b['Torque'] = features_df_b['Torque'].fillna(Torque_mode)
features_df_b['Seats'] = features_df_b['Seats'].fillna(Seats_mode)
```

```python
features_df_b
```

```python
features_df_b.isnull().sum()
```

```python
merged_df_bangalore = pd.concat([bangalore, features_df_b], axis=1)
```

```python
merged_df_bangalore.drop(['Top specification','Detailed engine'],axis = 1 ,inplace = True)
```

```python
merged_df_bangalore
```

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_features:
    contingency_table = pd.crosstab(car_df['price'], car_df[feature])
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)
    
    if p_value < 0.05:
        chi2_test.append('Reject Null Hypothesis')  
    else:
        chi2_test.append('Fail to Reject Null Hypothesis') 

test_result = pd.DataFrame({
    'Categorical Features': categorical_features,
    'Hypothesis Result': chi2_test
})
print(test_result)
```
